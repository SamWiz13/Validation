```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import pathlib
import matplotlib.pyplot as plt
import cv2
import cvlib as cv
```

```python
NUM_WORD =10000
(train_data, train_labels), (test_data, test_labels) = keras.datasets.imdb.load_data(num_words =NUM_WORD)
```

```python
def OneHotEncoding (sequence, dimension):
    results = np.zeros((len(sequence), dimension))
    for i, j in enumerate(sequence):
        results[i,j]=1.0
    return  results

train_data =OneHotEncoding(train_data, dimension =NUM_WORD)
test_data =OneHotEncoding(test_data, dimension =NUM_WORD)
```

# small_model

```python
small_model =keras.Sequential([
    keras.layers.Dense(4, activation =tf.nn.relu, input_shape =(NUM_WORD,)),
    keras.layers.Dense(4, activation =tf.nn.relu),
    keras.layers.Dense(1, activation =tf.nn.sigmoid),
])

small_model.compile(optimizer ='adam', loss ='binary_crossentropy', metrics =['accuracy','binary_crossentropy'])

small_model.summary()
```

```python
small_history =small_model.fit(train_data,
                                train_labels,
                                epochs =10,
                                batch_size =512,
                                validation_data =(test_data, test_labels),
                                verbose =2)
```

```python
small_model.evaluate(test_data, test_labels)
```

# 

```python
middle_model =keras.Sequential([
    keras.layers.Dense(16, activation =tf.nn.relu, input_shape =(NUM_WORD,)),
    keras.layers.Dense(16, activation =tf.nn.relu),
    keras.layers.Dense(1, activation =tf.nn.sigmoid),
])

middle_model.compile(optimizer ='adam', loss ='binary_crossentropy', metrics =['accuracy','binary_crossentropy'])

middle_model.summary()
```

```python
middle_history =middle_model.fit(train_data,
                                train_labels,
                                epochs =10,
                                batch_size =512,
                                validation_data =(test_data, test_labels),
                                verbose =2)
```

```python
middle_model.evaluate(test_data, test_labels)
```

```python
bigger_model =keras.Sequential([
    keras.layers.Dense(512, activation =tf.nn.relu, input_shape =(NUM_WORD,)),
    keras.layers.Dense(512, activation =tf.nn.relu),
    keras.layers.Dense(1, activation =tf.nn.sigmoid),
])

bigger_model.compile(optimizer ='adam', loss ='binary_crossentropy', metrics =['accuracy','binary_crossentropy'])

bigger_model.summary()
```

```python
bigger_history =bigger_model.fit(train_data,
                                train_labels,
                                epochs =10,
                                batch_size =512,
                                validation_data =(test_data, test_labels),
                                verbose =2)
```

```python
bigger_model.evaluate(test_data, test_labels)
```

```python
def plot_history(histories, key ='binary_crossentropy'):
    plt.figure(figsize =(10,10))
    
    for name,history in histories:
        value =plt.plot(history.epoch, history.history['val_' +key],
                       '--', label =name.title() +'val')
        plt.plot(history.epoch, history.history[key], color =value[0].get_color(),
                label =name.title()+ ' Train')
        
    plt.xlabel("Epochs")
    plt.ylabel(key.replace("_"," ").title())
    plt.legend()
    
    plt.xlim([0,max(history.epoch)])
    
plot_history([('small', small_history),
             ('middle', middle_history),
             ('bigger', bigger_history)])
```

```python
regulize_model =keras.models.Sequential([
    keras.layers.Dense(16, kernel_regularizer =keras.regularizers.l2(0.001), 
                       activation =tf.nn.relu, input_shape =(NUM_WORD,)),
    keras.layers.Dense(16, kernel_regularizer =keras.regularizers.l2(0.001),
                       activation =tf.nn.relu),
    keras.layers.Dense(1, activation =tf.nn.sigmoid),
])

regulize_model.compile(optimizer ='adam', 
                       loss ='binary_crossentropy', 
                       metrics =['accuracy','binary_crossentropy'])

regulize_history =regulize_model.fit(train_data,
                                train_labels,
                                epochs =10,
                                batch_size =512,
                                validation_data =(test_data, test_labels),
                                verbose =2)
```

```python
regulize_model.evaluate(test_data, test_labels)
```

```python
plot_history([('middle', middle_history),
             ('regulize', regulize_history)])
```

```python
dtp_model =keras.models.Sequential([
    keras.layers.Dense(16, activation =tf.nn.relu, input_shape =(NUM_WORD,)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(16, activation =tf.nn.relu),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation =tf.nn.sigmoid),
])

dtp_model.compile(optimizer ='adam', 
                  loss ='binary_crossentropy', 
                  metrics =['accuracy','binary_crossentropy'])

dtp_history =dtp_model.fit(train_data,train_labels,
                                epochs =10,
                                batch_size =512,
                                validation_data =(test_data, test_labels),
                                verbose =2)

```

```python
dtp_model.evaluate(test_data, test_labels)
```

```python
plot_history([('middle', middle_history),
             ('dtp', dtp_history)])
```
